```python
import sympy as sp
import random
import inspect
import json
import os
import numpy as np

# Function to generate a random expression with a given number of terms
def generate_random_augmented_expression(variables, num_terms=3, max_coeff=10, max_power=3):
    expr = 1
    for _ in range(num_terms):
        coeff = random.randint(1, max_coeff)
        term = coeff
        for var in variables:
            if random.random() > 0.5:
                power = random.randint(1, max_power)
                term *= var**power
        expr *= term

    # Adding complexity by adding another expression
    aug_expr = expr - sp.Rational(random.randint(1, 20), random.randint(1, 5))
    aug_expr *= sp.sin(variables[0] * random.randint(1, 10))
    return aug_expr

# Function to generate random test cases for the given expression
def generate_test_cases(expr, variables, num_cases=5):
    test_cases = []
    for _ in range(num_cases):
        input_values = {str(var): random.uniform(-10, 10) for var in variables}
        output_value = expr.subs(input_values)
        # Convert SymPy Float to Python float for JSON serialization
        test_cases.append({"input": input_values, "output": float(output_value)})
    return test_cases

# Function to determine complexity based on the number of terms and powers
def calculate_complexity(expr):
    complexity = len(expr.args)  # Number of terms
    for arg in expr.args:
        if isinstance(arg, sp.Pow):
            complexity += arg.exp  # Add powers
    return "low" if complexity <= 5 else "medium" if complexity <= 10 else "high"

# Initialize symbols
w, x, y, z = sp.symbols('w x y z')
variables = [w, x, y, z]

# Generate random expressions and their LaTeX representations
data = []
for i in range(10):
    random_expr = generate_random_augmented_expression(variables)
    latex_expr = sp.latex(random_expr)

    # Simplify the expression and convert to executable Python code
    # Generate Python function code for the solution

    solution_code = f"import numpy as np\nfrom numpy import sin, cos, tan\ndef augmented_equation({', '.join([str(var) for var in variables])}):\n"
    solution_code += f"    return {random_expr}\n"

    # Generate test cases
    test_cases = generate_test_cases(random_expr, variables)

    # Calculate complexity
    complexity = calculate_complexity(random_expr)

    # Create an entry with the required schema
    entry = {
        "task_id":  str(uuid.uuid4().hex),
        "sympy_exp": str(random_expr),
        "latex_exp": latex_expr,
        "solution": str(solution_code),
        "simplified_solution": "",
        "synthetic": True,
        "domain": "[-10, 10]",  # Domain set based on test case generation
        "test_cases": test_cases,
        "complexity": complexity,
        "equation_type": "Augmented",  # Assuming multivariable for these examples
        "output_type": "real"  # Assuming real values for these examples
    }

    data.append(entry)

# Check if the file exists and has content
# File path for JSON data on drive
file_path = '/content/drive/MyDrive/Colab Notebooks/augmented.json'

# Check if the file exists and has content
if os.path.exists(file_path) and os.stat(file_path).st_size > 0:
    # Save data into the JSON file
    with open(file_path, 'r+') as f:
        existing_data = json.load(f)
        existing_data.extend(data)
        f.seek(0)
        json.dump(existing_data, f, indent=4)
else:
    # Create a new file and save
    with open(file_path, 'w') as f:
        json.dump(data, f, indent=4)


print(f"Data saved to {file_path}")
```

```python
def generate_random_augmented_expression(variables, num_terms=3, max_coeff=10, max_power=3):
    expr_types = ['geometric', 'multivariable', 'simple_algebraic', 'exponential', 'fractional', 'trigonometric']

    # Randomly choose two different expression types
    expr_type1, expr_type2 = random.sample(expr_types, 2)

    def generate_expression(expr_type):
        expr = 0
        if expr_type == 'geometric':
            # Generate a geometric progression term
            common_ratio = random.randint(1, max_coeff)
            term = 1
            for i in range(num_terms):
                term *= (common_ratio * variables[i % len(variables)])
            expr = term

        elif expr_type == 'multivariable':
            # Generate a multivariable polynomial
            expr = 1
            for _ in range(num_terms):
                coeff = random.randint(1, max_coeff)
                term = coeff
                for var in variables:
                    if random.random() > 0.5:
                        power = random.randint(1, max_power)
                        term *= var**power
                expr *= term

        elif expr_type == 'simple_algebraic':
            # Generate a simple algebraic equation
            coeff = random.randint(1, max_coeff)
            expr = coeff * (variables[0] + random.randint(1, max_coeff))

        elif expr_type == 'exponential':
            # Generate an exponential equation
            base = random.randint(1, max_coeff)
            expr = base**variables[0]

        elif expr_type == 'fractional':
            # Generate a fractional equation
            numerator = random.randint(1, max_coeff) * variables[0]
            denominator = random.randint(1, max_coeff) * variables[1]
            expr = numerator / denominator

        elif expr_type == 'trigonometric':
            # Generate a trigonometric equation
            coeff = random.randint(1, max_coeff)
            expr = coeff * sp.sin(variables[0]) + sp.cos(variables[1] * random.randint(1, max_coeff))

        return expr

    # Generate the two expressions
    expr1 = generate_expression(expr_type1)
    expr2 = generate_expression(expr_type2)

    # Add the two expressions together
    aug_expr = expr1 + expr2

    return aug_expr
```

```python
# @title generate_algebraic(), generate_diophantine(), generate_exponential(), generate_exponential_decay() functions
import sympy as sp
import random
import inspect
import json
import os
import numpy as np
import uuid


def generate_algebraic():
    local_syms = sp.symbols('x y z')
    local_sym = random.choice(local_syms)
    local_coeffs = [random.randint(-10, 15) for _ in range(20)]
    while all(coeff == 0 for coeff in local_coeffs):
        local_coeffs = [random.randint(-10, 15) for _ in range(20)]
    return sum(random.choice(local_coeffs) * local_sym**i for i in range(random.randint(4,9))) + random.randint(0, 20)


def generate_diophantine():
    local_syms = random.choice([sp.symbols('v w x y z'), sp.symbols('a b c')])
    local_coeffs = [random.randint(-10, 15) for _ in range(20)]
    while all(coeff == 0 for coeff in local_coeffs):
        local_coeffs = [random.randint(-10, 15) for _ in range(20)]
    return sum(random.choice(local_coeffs) * random.choice(local_syms) for _ in range(random.randint(3,8))) + random.randint(0, 20)


def generate_exponential():
    local_syms = sp.symbols('a b c v w x y z')
    local_coeffs = [random.randint(-10, 15) for _ in range(20)]
    while all(coeff == 0 for coeff in local_coeffs):
        local_coeffs = [random.randint(-10, 15) for _ in range(20)]
    return sum(random.randint(2, 9) ** random.choice(local_syms) + random.randint(0, 20) + \
               random.choice(local_coeffs) * random.choice(local_syms) ** random.randint(1, 5) for _ in range(random.randint(2, 4)))


def generate_exponential_decay():
    local_syms = random.choice([sp.symbols('v w x y z'), sp.symbols('a b c')])
    local_coeffs = [random.uniform(-10, 10) for _ in range(20)]
    while all(coeff == 0 for coeff in local_coeffs):
        local_coeffs = [random.uniform(-10, 10) for _ in range(20)]
    return sum(random.choice(local_coeffs) * random.choice(local_syms) ** random.randint(1, 4) for _ in range(random.randint(2, 4))) + \
           random.choice(local_coeffs) * sp.exp(random.choice(local_coeffs)*random.choice(local_syms))
```

```python
# @title generate_logrithmic(), generate_multivariable(), generate_rational(), generate_summation(), generate_trigonometric() functions

def generate_logrithmic():
    local_syms = sp.symbols('v w x y z')
    local_sym = random.choice(local_syms)
    local_coeffs = [random.randint(-10, 15) for _ in range(20)]
    while all(coeff == 0 for coeff in local_coeffs):
        local_coeffs = [random.randint(-10, 15) for _ in range(20)]
    return random.choice(local_coeffs) * sp.log(random.choice(local_coeffs)*local_sym) + random.choice(local_coeffs)


def generate_multivariable():
    local_syms = sp.symbols('v w x y z')
    local_coeffs = [random.randint(-10, 15) for _ in range(20)]
    while all(coeff == 0 for coeff in local_coeffs):
        local_coeffs = [random.randint(-10, 15) for _ in range(20)]
    return sum(random.choice(local_coeffs) * random.choice(local_syms)**random.randint(1, 5) for _ in range(random.randint(3, 10)))

def generate_rational():
    local_syms = random.choice([sp.symbols('v w x y z'), sp.symbols('a b c')])
    terms = []
    for _ in range(random.randint(1, 2)):
        num = random.choice([
            random.randint(-10, 10) * random.choice(local_syms),
            random.randint(-10, 10) * random.choice(local_syms) + random.randint(-10, 10),
            random.randint(-10, 10) * random.choice(local_syms)**random.randint(2, 3),
            random.randint(-10, 10) * random.choice(local_syms)**random.randint(2, 3) + random.randint(-10, 10) * random.choice(local_syms)
        ])
        den = random.randint(-10, 10) * random.choice(local_syms)**random.randint(1, 3) + random.randint(-10, 10)
        terms.append(num / den)
    return sum(terms) + random.randint(-10, 10)

def generate_summation():
    local_syms = sp.symbols('a b c i n v w x y z')
    local_sym = random.choice(local_syms)
    return sp.Sum(random.randint(3,9)**local_sym, (local_sym, 1, random.randint(2, 12)))

def generate_trigonometric():
    local_syms = sp.symbols('w v x y z theta alpha beta gamma phi')
    local_sym = random.choice(local_syms)
    trig_funcs = [sp.sin, sp.cos, sp.tan, sp.csc, sp.sec, sp.cot, sp.exp]
    trig_func = random.choice(trig_funcs)
    return random.randint(1, 9) * trig_func(random.randint(1, 12) * local_sym) + random.randint(1, 15)
```

```python
# @title generate_expression() function

def generate_expression(expr_type):
    if expr_type == 'algebraic':
        return generate_algebraic()
    elif expr_type == 'diophantine':
        return generate_diophantine()
    elif expr_type == 'exponential':
        return generate_exponential()
    elif expr_type == 'exponential_decay':
        return generate_exponential_decay()
    elif expr_type == 'geometric':
        return generate_geometric()
    elif expr_type == 'logrithmic':
        return generate_logrithmic()
    elif expr_type == 'multivariable':
        return generate_multivariable()
    elif expr_type == 'rational':
        return generate_rational()
    elif expr_type == 'summation':
        return generate_summation()
    elif expr_type == 'trigonometric':
        return generate_trigonometric()
    else:
        raise ValueError(f"Unknown expression type: {expr_type}")
```

```python
#@title generate actual data
data = []
num_exprs = 9000
from tqdm import tqdm
expr_types = ['algebraic', 'diophantine', 'exponential', 'exponential_decay', 'geometric', 'logrithmic', 'multivariable', 'rational', 'summation', 'trigonometric']

for i in tqdm(range(num_exprs), desc="Generating data"):
  # Step 1: Generate sympy expression
    chosen_types = random.sample(expr_types, 4)
    chosen_expressions = [generate_expression(expr_type) for expr_type in chosen_types]
    e1, e2, e3, e4 = chosen_expressions
    combinations = [
        lambda: (e1 + e2) * (e3 + e4),
        lambda: (e1 - e2) * (e3 - e4),
        lambda: (e1 * e2) + (e3 * e4),
        lambda: (e1 / e2) + (e3 / e4),
        lambda: ((e1 + e2) * e3) / e4,
        lambda: ((e1 - e2) * e3) / e4,
        lambda: (e1 * e2) / (e3 + e4),
        lambda: (e1 * e2) / (e3 - e4),
        lambda: e1 - (e2 * e3 / e4),
        lambda: (e1 + e2 + e3) / e4,
        lambda: (e1 / e2) * (e3 / e4),
        lambda: ((e1 + e2) / e3) - e4,
        lambda: (e1 * e2 - e3) / e4,
        lambda: e1 / (e2 + e3 - e4)
    ]
    section = num_exprs // (len(combinations)-1)
    sympy_expr = combinations[i//section]()

  # Step 2: Generate python function
    # extracting symbols used in sympy expression
    symbol_list = sympy_expr.free_symbols

    # Remove cmath from the solution code as it is not needed
    solution_code = f"import numpy as np\nfrom sympy import *\ndef augmented_equation({', '.join([str(var) for var in symbol_list])}):\n    return {sympy_expr}\n"
    solution_code.replace("pi", "np.pi").replace("np.np.pi", "np.pi")
    solution_code = re.sub(r"Sum\((.*?), \((\w+), (\d+), (\d+)\)\)", r"sum(\1 for \2 in range(\3, \4 + 1))", solution_code)
  # Step 3: Generate data
    # Create an entry with the required schema
    entry = {
        "task_id":  str(uuid.uuid4().hex[:8]),
        "sympy_exp": str(sympy_expr),
        "latex_expression": sp.latex(sympy_expr),
        "solution": str(solution_code),
        "simplified_solution": "",
        "synthetic": True,
        "domain": "[-10, 10]",  # Domain set based on test case generation
        "test_cases": [],
        "complexity": random.randint(1,5),
        "equation_type": "augmented_equation",  # Assuming multivariable for these examples
        "output_type": ("complex" if 'log' in str(sympy_expr) else "real")
    }

    data.append(entry)

print("Data Generated!")
```

```python
# @title evaluate_code function
import json
import re
import math
import csv
import torch

def evaluate_code(generated_code: str, test_cases: list):
    import re
    output_list = []
    error_list = [str(float('inf'))]*len(test_cases)
    is_solution_valid = True

    # Step 1: Extract and execute import statements
    # import_pattern = re.compile(r'^\s*(import\s+\w+(?:\s+as\s+\w+)?|from\s+\w+\s+import\s+\w+)', re.MULTILINE)
    import_pattern = re.compile(r'^\s*(import\s+\w+(?:\s+as\s+\w+)?|from\s+\w+\s+import\s+\*|from\s+\w+\s+import\s+\w+(?:,\s*\w+)*)', re.MULTILINE)
    imports = import_pattern.findall(generated_code)

    for import_statement in imports:
        try:
            exec(import_statement, globals())
        except ImportError as e:
            print(f"ImportError in executing import statement: {import_statement}. Error: {e}")
        except SyntaxError as e:
            print(f"SyntaxError in import statement: {import_statement}. Error: {e}")

    # Remove import statements from generated_code to isolate the function definition
    generated_code = import_pattern.sub('', generated_code)

    # Step 2: Extract the function definition
    function_pattern = re.compile(r'(def\s+\w+\([^)]*\):(?:\n\s+.+)+)')
    match = function_pattern.search(generated_code)
    if match:
        function_def = match.group(1)
    else:
        print(f"No valid function definition found!")
        is_solution_valid = False
        return error_list, is_solution_valid

    # Step 3: Extract the function parameters
    param_pattern = re.compile(r'def\s+(\w+)\(([^)]*)\)')
    param_match = param_pattern.search(function_def)
    if param_match:
        function_name = param_match.group(1)
        parameters = param_match.group(2).split(', ')
        # print(f"Parameters: {parameters}")
    else:
        print(f"Couldn't extract function parameters!")
        is_solution_valid = False
        return error_list, is_solution_valid

    # Step 4: Execute the function definition
    local_scope = {}
    try:
        exec(function_def, globals(), local_scope)
    except SyntaxError as e:
        print(f"SyntaxError in function definition!")

    # Ensure the function is defined in local_scope
    if function_name not in local_scope:
        print(f"Function '{function_name}' not found after exec!")
        is_solution_valid = False
        return error_list, is_solution_valid

    # Step 5: Test each case in the test_cases
    is_result_valid_list = []
    for test_case in test_cases:
        is_result_valid = True
        input_values = test_case["input"]
        try:
            function_args = {param: input_values.get(param, 0) for param in parameters}
        except Exception as e:
            is_result_valid = False
            print(f"Error finding arguments: {e}")
            result = float('inf')
            output_list.append(str(result))
            is_result_valid_list.append(is_result_valid)
            continue
        function_to_call = local_scope[function_name]

        try:
            result = function_to_call(**function_args)
        except Exception as e:
            # print(f"Function arguments: {function_args}")
            if str(e) != "math range error":
              print(f"An error occurred while calculating result: {e}")
              is_result_valid = False
            else:
              is_result_valid = True
            result = float('inf')
            output_list.append(str(result))
            is_result_valid_list.append(is_result_valid)
            continue

        # Convert result to a Python number if it's a tensor
        if isinstance(result, torch.Tensor):
            result = result.item()
        elif isinstance(result, (int, float)):
            result = float(result)
        elif hasattr(result, 'evalf'):
            result = result.evalf()

        # Format the complex number to the desired string format
        if result != float('inf'):
            result = complex(result)
            if result.imag:
                if result.imag > 0:
                    result = f'{result.real:.6f}+{result.imag:.6f}j'
                else:
                    result = f'{result.real:.6f}-{abs(result.imag):.6f}j'
            else:
                result = f'{result.real:.6f}'

        # Append the result to the output_list
        output_list.append(str(result))
        is_result_valid_list.append(is_result_valid)

    return output_list, all(is_result_valid_list)
```
